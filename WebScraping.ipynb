```python
import undetected_chromedriver as uc
import pyautogui
import time
from io import StringIO
import pandas as pd

# Configuração do Driver (usando undetected para evitar o bloqueio inicial)
options = uc.ChromeOptions()
driver = uc.Chrome(options=options)
driver.maximize_window() # OBRIGATÓRIO: para o botão estar sempre no mesmo lugar

url = "https://www.transfermarkt.com/sport-club-corinthians-paulista/spielplan/verein/199/saison_id/2022/plus/1#BRA1"
driver.get(url)

# Espera a página carregar e o pop-up aparecer
print("⏳ Aguardando 10 segundos para o pop-up carregar...")
time.sleep(10)

# ==========================================
# ESTRATÉGIA PYAUTOGUI
# ==========================================

# Opção A: Clique por coordenada (Pode variar dependendo da sua resolução)
# No meu monitor 1920x1080, o botão "Accept" costuma ficar no meio/baixo.
# Você pode testar as coordenadas usando pyautogui.position()
print("🖱️ Tentando clicar no botão de Cookies...")
pyautogui.click(x=960, y=800) # Ajuste esses números se não clicar no local certo

# Opção B (Mais segura): Pressionar 'Enter' 
# Geralmente o botão de aceite já vem focado ou aceita o comando de teclado
pyautogui.press('enter')

time.sleep(5) # Espera o overlay sumir

# ==========================================
# EXTRAÇÃO DOS DADOS
# ==========================================
try:
    html_page = driver.page_source
    tabelas = pd.read_html(StringIO(html_page))
    
    for df in tabelas:
        if "Date" in df.columns:
            print("✅ Tabela capturada com sucesso!")
            df.to_csv("jogos_corinthians_2023.csv", index=False)
            break
finally:
    driver.quit()
```
