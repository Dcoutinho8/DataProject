```python
# ==========================================
# ⚽ 2022
# ==========================================

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from io import StringIO
import pandas as pd
from sqlalchemy import create_engine

# ==========================================
# 🔑 CONFIG BANCO
# ==========================================
usuario = 'postgres'
senha = "danilo123"
host = "localhost"
porta = "5432"
banco = "postgres"

engine = create_engine(f"postgresql://{usuario}:{senha}@{host}:{porta}/{banco}")

# ==========================================
# 🚗 CONFIG DRIVER
# ==========================================
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

# ==========================================
# 🌐 URL (SEM FILTRO!)
# ==========================================
url = "https://www.transfermarkt.com/sport-club-corinthians-paulista/spielplan/verein/199/plus/1?saison_id=2021"
driver.get(url)

# ==========================================
# 🍪 COOKIES
# ==========================================
try:
    botao_cookie = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[contains(text(),"Accept")]'))
    )
    botao_cookie.click()
    print("✅ Cookies aceitos!")
except:
    print("⚠️ Cookie não apareceu")

# ==========================================
# 📊 EXTRAÇÃO
# ==========================================
try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, "table"))
    )

    html_page = driver.page_source
    tabelas = pd.read_html(StringIO(html_page))

    dfs_validos = []

    for df in tabelas:
        if "Date" in df.columns:
            dfs_validos.append(df)

    if dfs_validos:
        df_final = pd.concat(dfs_validos, ignore_index=True)

        print(f"✅ {len(dfs_validos)} tabelas encontradas!")
        print(f"✅ {len(df_final)} jogos no total!")

        df_final.to_sql(
            name="jogos_corinthians_2022",
            con=engine,
            if_exists="replace",
            index=False
        )

        print("✅ Todos os jogos salvos no PostgreSQL!")
    else:
        print("❌ Nenhuma tabela válida encontrada!")

finally:
    driver.quit()
```
