```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
# Removi o webdriver_manager temporariamente para evitar o erro de rede
import pandas as pd
import time
import io # Necessário para o novo padrão do Pandas

# Tente iniciar o Chrome de forma simples
# Se você tiver o Chrome instalado, o Selenium 4.6+ gerencia isso sozinho
driver = webdriver.Chrome() 

url = 'https://www.transfermarkt.com/sport-club-corinthians-paulista/spielplan/verein/199/saison_id/2022/plus/1#BRA1'

try:
    driver.get(url)
    time.sleep(5) # Aguarda o carregamento

    html_page = driver.page_source

    # Corrigindo o erro do Pandas usando StringIO
    # Isso remove o aviso de "FutureWarning"
    tabelas = pd.read_html(io.StringIO(html_page))

    found = False
    for df in tabelas:
        # Verificamos se a tabela tem as colunas que nos interessam
        if any('Data' in str(col) for col in df.columns):
            print("Tabela encontrada com sucesso!")
            print(df.head())
            df.to_csv('jogos_corinthians.csv', index=False)
            found = True
            break
    
    if not found:
        print("Não foi possível encontrar a tabela de jogos específica.")

finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import service
from webdriver_manager.chrome import ChromeDriverManager
import os
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import re



driver = webdriver.Chrome()

url = 'https://www.transfermarkt.com/sport-club-corinthians-paulista/spielplan/verein/199/saison_id/2022/plus/1#BRA1'

try:
    driver.get(url)
    # Espera um pouco para a página carregar (o ideal é usar WebDriverWait)
    time.sleep(5)

    # 2. Em vez de Select, vamos pegar o conteúdo HTML da página
    html_page = driver.page_source

    # 3. O Pandas consegue ler todas as tabelas do HTML de uma vez
    tabelas = pd.read_html(html_page)

    # No Transfermarkt, a tabela de jogos geralmente é uma das maiores
    # Vamos procurar a tabela que contém "Data" ou "Resultado"
    for df in tabelas:
        if 'Data' in df.columns or 'Resultado' in df.columns:
            print("Tabela encontrada!")
            print(df.head())
            # Salvar em CSV
            df.to_csv('jogos_corinthians.csv', index=False)
            break

finally:
    driver.quit()
```
