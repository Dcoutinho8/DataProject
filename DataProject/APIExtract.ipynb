```python
import requests
import pandas as pd
import os # <-- Faltava importar o módulo do sistema
from dotenv import load_dotenv # <-- Faltava importar a função para ler o .env

# 1. Carrega as variáveis do arquivo .env
load_dotenv()

# 2. Pega a chave do ambiente (certifique-se que no .env esteja API_KEY=suachave)
api_key = os.getenv('API_KEY')

url = "https://v3.football.api-sports.io/players"

# 3. Verifica se a chave foi carregada antes de continuar
if api_key:
    print("Chave carregada com sucesso.")
else:
    print("Erro: API_KEY não encontrada no arquivo .env")
    exit() # Para o código se não tiver a chave

# 4. COLOCA A VARIÁVEL AQUI NOS HEADERS
headers = {
    "x-apisports-key": api_key # <-- Aqui você usa a variável que guardou a chave
}

seasons = [2022, 2023, 2024]
dfs = []

for season in seasons:
    params = {
        "team": 131,
        "season": season
    }

    print(f"Buscando dados da temporada {season}...")

    response = requests.get(
        url,
        headers=headers,
        params=params,
        proxies={
            "http": "http://internet.ford.com:83",
            "https": "http://internet.ford.com:83"
        }
    )

    
    if response.status_code == 200:
        data = response.json()
    
        if "response" in data and data["response"]:
            df_temp = pd.json_normalize(data["response"])
            df_temp["season"] = season
            dfs.append(df_temp)
        else:
            print(f"Nenhum dado encontrado para a temporada {season}")
    else:
        print(f"Erro na requisição ({season}): {response.status_code}")

# Só tenta concatenar se a lista dfs não estiver vazia
if dfs:
    df = pd.concat(dfs, ignore_index=True)
    print(df.head())
    print("Total de jogadores:", len(df))
    df.to_csv("corinthians_matches_jogadores2022_2024.csv", index=False)
else:
    print("Nenhum dado foi coletado.")
```
